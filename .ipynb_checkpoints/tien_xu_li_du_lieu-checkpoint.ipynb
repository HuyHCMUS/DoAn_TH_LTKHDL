## Import các thư viện cần thiết

```python
import pandas as pd
import numpy as np
```

## Đọc file dữ liệu

Dữ liệu của chúng ta được lưu trong file 'Anime.csv'. Chúng ta sẽ đọc dữ liệu và lưu vào DataFrame `data_df`

```python
data_df = pd.read_csv('Anime.csv')
data_df.head(35)
```

Hiển thị 5 dòng đầu của dữ liệu để xem đọc đúng chưa

Như đã nói ở phần khám phá dữ liệu, để dễ dàng phân tích cũng như trích xuất ngày, tháng, năm ở 2 cột này, chúng ta nên đưa 2 cột này về kiểu dữ liệu datetime.

```python
data_df['Start_Aired'] = pd.to_datetime(data_df['Start_Aired'], format='%b %d, %Y', errors='coerce')
```
